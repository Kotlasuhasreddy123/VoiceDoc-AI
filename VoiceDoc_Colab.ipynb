## Step 3: Initialize System

```python
import sys
sys.path.append('/content/VoiceDoc-AI')

from voicedoc_core import VoiceDocCore
from ml_engine import MedicalTriageEngine
from medical_knowledge_graph import MedicalKnowledgeGraph
from rag_engine import MedicalRAG
from clinical_decision_support import ClinicalDecisionSupport
from voice_biomarker_analyzer import VoiceBiomarkerAnalyzer
import numpy as np
import json

# Initialize main system (use_gemma=False = mock mode, no GPU needed)
voicedoc = VoiceDocCore(use_gemma=False)
print('VoiceDoc AI initialized successfully')
print('All 7 features loaded')
```

## Step 4: Batch Triage Test (All 5 Cases)

```python
import pandas as pd

test_cases = [
    ('Mild Headache',      'I have a mild headache',                         1, False),
    ('Fever + Cough',      'High fever and cough for 2 days',                2, True),
    ('Severe Chest Pain',  'Severe chest pain and difficulty breathing',     1, False),
    ('Deep Wound',         'Deep cut on my leg with bleeding',               1, False),
    ('Nausea + Vomiting',  'Nausea and vomiting for 4 hours',               1, False),
]

emoji = {'green': '🟢', 'yellow': '🟡', 'red': '🔴'}
rows = []
for name, symptoms, duration, fever in test_cases:
    r = voicedoc.triage_with_function_calling(symptoms, duration, fever)
    rows.append({
        'Test Case': name,
        'Triage': f"{emoji[r.triage_level]} {r.triage_level.upper()}",
        'Risk Score': f"{r.risk_score:.1f}/100",
        'Confidence': f"{r.confidence*100:.0f}%"
    })

df = pd.DataFrame(rows)
print('Batch Triage Results')
print('=' * 65)
print(df.to_string(index=False))
print()
print('Severe Chest Pain correctly classified as RED (risk 96/100)')
print('Mild Headache correctly classified as GREEN (risk 25/100)')
```

## Step 5: Medical Knowledge Graph (SNOMED CT)

```python
# Get clinical context for a set of symptoms
symptoms = ['fever', 'cough', 'chest_pain']
context = voicedoc.get_clinical_context(symptoms)

print('Medical Knowledge Graph — Clinical Context')
print('=' * 55)
print(f"Symptoms: {', '.join(symptoms)}")
print(f"Urgency Level: {context['urgency_level'].upper()}")
print(f"SNOMED CT Codes: {', '.join(context['snomed_codes'])}")
print(f"ICD-11 Codes: {', '.join(context['icd11_codes'])}")
print()
print('Top Differential Diagnoses:')
for dx in context['differential_diagnoses'][:3]:
    print(f"  • {dx['diagnosis']}: {dx['confidence']:.0%} confidence")
print()
print('Clinical Guidelines:')
for g in context['clinical_guidelines'][:3]:
    print(f"  • {g}")
```

## Step 6: Voice Biomarker Analysis

```python
# Simulate voice biomarker analysis with synthetic audio
analyzer = VoiceBiomarkerAnalyzer()
sr = 16000

# Simulate a stressed voice (higher energy variance)
audio_stressed = np.random.randn(sr) * 0.15
features = analyzer.extract_acoustic_features(audio_stressed, sr)
signals = analyzer.detect_health_signals(features)

print('Voice Biomarker Analysis')
print('=' * 50)
print('Acoustic Features (key):')
print(f"  Pitch Mean:      {features['pitch_mean']:.1f} Hz")
print(f"  Pitch Variance:  {features['pitch_variance']:.4f}")
print(f"  Energy Mean:     {features['energy_mean']:.4f}")
print(f"  Speech Rate:     {features['speech_rate']:.0f} words/min")
print()
print('Detected Health Signals:')
for name, data in signals.items():
    if isinstance(data, dict) and 'detected' in data:
        status = 'DETECTED' if data['detected'] else 'Not detected'
        conf = data.get('confidence', 0)
        print(f"  {name:<25} {status}  ({conf:.0%})")
mh = signals.get('mental_health_risk', {})
print()
print(f"Mental Health Risk: {mh.get('risk_level', 'unknown').upper()}")
print(f"Recommendation: {mh.get('recommendation', 'N/A')}")
```

## Step 7: RAG — Evidence-Grounded Recommendations

```python
rag = MedicalRAG()

symptoms = ['fever', 'cough']
triage_result = {'triage_level': 'yellow', 'risk_score': 49}
summary = rag.generate_rag_summary(symptoms, triage_result)

print('RAG — Evidence-Grounded Recommendations')
print('=' * 55)
print(f"Grounding Confidence: {summary['grounding_confidence']:.0%}")
print(f"Validation: {summary['validation']['validation_status']}")
print()
print('Evidence-Based Recommendations:')
for rec in summary['evidence_based_recommendations'][:4]:
    print(f"  • {rec['recommendation']}")
    print(f"    Source: {rec['source']}")
print()
print('Retrieved Evidence Sources:')
for ev in summary['retrieved_evidence']:
    print(f"  • {ev['title']} ({ev['source']})")
```

## Step 8: Clinical Decision Support — Audit Trail

```python
cds = ClinicalDecisionSupport()

patient_data = {
    'patient_id': 'P001',
    'symptoms': ['fever', 'cough'],
    'fever': True,
    'fever_temperature': 39.2,
    'duration_days': 3
}
triage_result = {'triage_level': 'yellow', 'risk_score': 49, 'confidence': 0.65}

report = cds.generate_cds_report(triage_result, patient_data)

print('Clinical Decision Support Report')
print('=' * 55)
print(f"Patient: {report['patient_id']}")
print(f"Triage: {report['triage_decision']['level'].upper()}")
print(f"Risk Score: {report['triage_decision']['risk_score']}/100")
print()
print(f"Rules Triggered: {report['decision_support']['triggered_rules']}")
for rule in report['decision_support']['rules']:
    print(f"  • {rule['rule_display_name']}")
    print(f"    Recommendation: {rule['recommendation']}")
print()
print('Compliance:')
for k, v in report['compliance'].items():
    print(f"  {'✓' if v else '✗'} {k}")
```

## Step 9: Full Integrated Analysis

```python
print('INTEGRATED ANALYSIS — All 7 Features')
print('=' * 60)

symptoms_text = 'High fever, persistent cough, and chest pain for 3 days'
symptoms_list = ['fever', 'cough', 'chest_pain']
print(f'Patient: "{symptoms_text}"\n')

# 1. Triage
triage = voicedoc.triage_with_function_calling(symptoms_text, duration_days=3, has_fever=True)
print(f'1. TRIAGE DECISION')
print(f'   Level: {triage.triage_level.upper()}  |  Risk: {triage.risk_score}/100  |  Confidence: {triage.confidence*100:.0f}%')

# 2. Clinical context
ctx = voicedoc.get_clinical_context(symptoms_list)
print(f'\n2. CLINICAL CONTEXT (SNOMED CT)')
print(f'   Urgency: {ctx["urgency_level"].upper()}')
print(f'   SNOMED Codes: {chr(44).join(ctx["snomed_codes"])}')
print(f'   Top Diagnosis: {ctx["differential_diagnoses"][0]["diagnosis"]} ({ctx["differential_diagnoses"][0]["confidence"]:.0%})')

# 3. RAG
rag_summary = rag.generate_rag_summary(symptoms_list, {'triage_level': triage.triage_level, 'risk_score': triage.risk_score})
print(f'\n3. RAG GROUNDING')
print(f'   Confidence: {rag_summary["grounding_confidence"]:.0%}')
print(f'   Top Recommendation: {rag_summary["evidence_based_recommendations"][0]["recommendation"]}')

# 4. CDS
cds_report = cds.generate_cds_report(
    {'triage_level': triage.triage_level, 'risk_score': triage.risk_score, 'confidence': triage.confidence},
    {'patient_id': 'P001', 'symptoms': symptoms_list, 'fever': True, 'fever_temperature': 39.5, 'duration_days': 3}
)
print(f'\n4. CLINICAL DECISION SUPPORT')
print(f'   Rules Triggered: {cds_report["decision_support"]["triggered_rules"]}')
print(f'   SNOMED Compliant: {cds_report["compliance"]["snomed_ct_compliant"]}')
print(f'   Audit Trail: Generated')

print(f'\n5. HOME CARE INSTRUCTIONS')
for i, inst in enumerate(triage.home_care_instructions, 1):
    print(f'   {i}. {inst}')

print(f'\n6. ESCALATION GUIDANCE')
print(f'   {triage.escalation_guidance}')

print(f'\n{"="*60}')
print('All 7 features working correctly.')
```
